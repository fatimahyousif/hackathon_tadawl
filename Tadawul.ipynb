## üíæ Dataset:


Our Columns:
1-symbol: Symbol or ticker of the stock.
2-name: Name of the stock.
3-trading_name: Trading name of the stock.
4-sector: Sector to which the stock belongs.
5-date: Date of the data.
6-open: Opening price of the stock on that date.
7-high: Highest price of the stock during the trading day.
8-low: Lowest price of the stock during the trading day.
9-close: Closing price of the stock on that date.
10-change: Change in the price compared to the previous day.
11-perc_Change: Percentage change in the price compared to the previous day.
12-volume_traded: Volume of stocks traded on that date.
13-value_traded: Value of stocks traded on that date.
14-no_trades: Number of trades that occurred on that date.

Hande null value by create a function called high_low_null that fills null values in the 'high' , 'low', no_trades and open attributes using the forward fill (ffill) method üéâ

```python
def high_low_null(df):
    # Fill null values in 'high', 'low', 'open', and 'no_trades' columns using forward fill
    df_Td['high'] = df_Td['high'].fillna(method='ffill')
    df_Td['low'] = df_Td['low'].fillna(method='ffill')
    df_Td['open'] = df_Td['open'].fillna(method='ffill')
    df_Td['no_trades '] = df_Td['no_trades '].fillna(method='ffill')
    return df


#  columns'high', 'low', 'open', and 'no_trades' containing null values
# Call the high_low_null function to fill null values
df = high_low_null(df_Td)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.pie(sector_counts, labels=sector_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Stocks Across Sectors')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

```
